## Hyperparameter tuning using RandomizedSearchcv


```python
# Create a hyperparameter grid for LogisticRegression
log_reg_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}
```

```python
# Tune LogisticRegression

np.random.seed(42)

# Setup random hyperparameter search for LogisticRegression
rs_log_reg = RandomizedSearchCV(LogisticRegression(),
                                param_distributions=log_reg_grid,
                                cv=5,
                                n_iter=20,
                                verbose=True)

# Fit random hyperparameter search model for LogisticRegression
rs_log_reg.fit(X_train, y_train)
```

```python
score = rs_log_reg.score(X_test,y_test)
score
```

## Using Grid Search CV Logistic Regression

```python
log_reg_grid = {'C': np.logspace(-4,4,30),
               "solver":["liblinear"]}

#setup  the gird cv
gs_log_reg = GridSearchCV(LogisticRegression(),
                          param_grid=log_reg_grid,
                          cv=5,
                          verbose=True)

#fit grid search cv
gs_log_reg.fit(X_train,y_train)
```

```python
gs_log_reg.score(X_test,y_test)
```

## Evaluating the tuned model by RandomizedSearchcv

```python
y_preds = gs_log_reg.predict(X_test)
y_preds
```

```python
np.array(y_test)
```

```python
# plot ROC curve
plot_roc_curve(gs_log_reg,X_test,y_test)
```

```python
#Confusion matrix
print(confusion_matrix(y_test,y_preds))
```

```python
sns.set(font_scale=1)
import seaborn as sns
sns.heatmap(confusion_matrix(y_test,y_preds), annot=True,cbar=False, fmt='g')
plt.xlabel("True label")
plt.ylabel("Predicted label");
```

```python
# Create a new classifier with best parameters
clf = LogisticRegression(C=2.592943797404667,solver="liblinear")
```

```python
# Cross-validated accuracy
cv_acc = cross_val_score(clf, X, y, cv=5, scoring="accuracy")
cv_acc
```

```python
cv_acc = np.mean(cv_acc)
cv_acc
```

```python
# Cross-validated precision
cv_precision = cross_val_score(clf, X,y,cv=10,scoring="precision")
cv_precision=np.mean(cv_precision)
cv_precision
```

```python
# Cross-validated recall
cv_recall = cross_val_score(clf,X, y,cv=10,scoring="recall")
cv_recall = np.mean(cv_recall)
cv_recall
```

```python
# Cross-validated f1-score
cv_f1 = cross_val_score(clf,
                         X,
                         y,
                         cv=10,
                         scoring="f1")
cv_f1 = np.mean(cv_f1)
cv_f1
```

```python
# Visualize cross-validated metrics
cv_metrics = pd.DataFrame({"Accuracy": cv_acc,
                           "Precision": cv_precision,
                           "Recall": cv_recall,
                           "F1": cv_f1},
                          index=[0])

cv_metrics.T.plot.bar(title="Cross-validated classification metrics",
                      legend=False);
```
